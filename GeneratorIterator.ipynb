```python
class RemoteControl:
  def __init__(self):
    self.channels = ["NETFLIX", "PRIME", "HBO", "CNN", "BBC"]
    self.index = -1

  def __next__(self):
    self.index += 1
    if self.index == len(self.channels):
      raise StopIteration
    return self.channels[self.index]

  def __iter__(self):
    return self

r = RemoteControl()
for channel in r:
  print(channel)
```
